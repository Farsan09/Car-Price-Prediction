```python
#Cleaning the Data
```

```python
df.head()
```

```python
df.info()
```

```python
#Creating a function to split the numbers from the letters to have an only integer column
```

```python
df.info()
```

```python
df.head()
```

```python
def nom (text):
    if isinstance(text,str):
        return text.split()[0]
    else: 
        return text
    
```

```python
df['Distance Driven'] = df['Mileage'].apply(nom)
```

```python
df['Engine Capacity'] = df['Engine'].apply(nom)
```

```python
df['Power'] = df['Max_power'].apply(nom)
```

```python
df.head()
```

```python
##Converting the new columns to Machine Learning Readable format
```

```python
df['Distance Driven'] = df['Distance Driven'].astype(float)
```

```python
df['Engine Capacity'] = df['Engine Capacity'].astype(float)
```

```python
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')
```

```python
df.info()
```

```python
##Filling missing values with the mean 
```

```python
df['Seats'] = df['Seats'].fillna(df['Seats'].mean())
```

```python
df['Distance Driven'] = df['Distance Driven'].fillna(df['Distance Driven'].mean())
```

```python
df['Engine Capacity'] = df['Engine Capacity'].fillna(df['Engine Capacity'].mean())
```

```python
df['Power'] = df['Power'].fillna(df['Power'].mean())
```

```python
df.head(2)
```

```python
df_2 = df.drop(['name', 'Mileage', 'Engine', 'Max_power','Torque/Rpm'],axis = 1)
```

```python
df_2.head()
```

```python
#Converting Categorical data to dummies
```

```python
final_data = pd.get_dummies(df_2, columns = ['Fuel','Seller_type','Transmission','Owner'], drop_first = True)
```

```python
final_data.head(10)
```

```python
final_data.info()
```

```python
#Test_Train_Split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('selling_price',axis = 1)
y = final_data['selling_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
#Intercept
print(lm.intercept_)
```

```python
#Scatter Plot
```

```python
plt.scatter(y_test,predictions)
plt.show()
```

```python
figure = plt.figure(figsize = (7,4))
sns.displot(y_test-predictions, kde = True)
plt.show()
```

```python
#Regression Reports
```

```python
from sklearn import metrics
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test,predictions))
```

```python
print('MSE: ', metrics.mean_squared_error(y_test,predictions))
```

```python
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
#COEFFICIENTS
```

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
